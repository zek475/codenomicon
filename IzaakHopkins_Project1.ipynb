>### Combat  
![Combat](https://i.ibb.co/8rSL4Y0/Screen-Shot-2021-04-15-at-8-53-17-PM.png)
The combat of this game is based on 'Rock, Paper, Scissors', with a few changes and upgrades...\
Similar to 'Rock, Paper, Scissors': ***MELEE beats MAGIC, MAGIC beats RANGED, RANGED beats MELEE***\
The player will pick their attack, and a random number will generate to decide the enemy's attack.
**But, If you both pick the same attack you CLASH** and the function tiebreaker() is called.\
![Tiebreaker](https://i.ibb.co/Pwps0LL/Screen-Shot-2021-04-15-at-9-48-00-PM.png)
**TIEBREAKER**: After clashing attacks with your enemy they become stunned, and you must then decide to either       **COUNTER or GUARD**. COUNTER will give you a chance to attack, but leave you vulnerable if your enemy doesn't stay stunned. GUARD will allow you to ensure that you take no damage, but you will lose the opportunity to counter-attack even if your enemy stays stunned. A random number is generated to decide if the enemy stays stunned.\
All of this is done using IF functions to check the randomly generated attacks from the enemy against the user's input for each round.\
The function checks many iterations of uppercase/lowercase options for each selection, and will also prompt the user again if the input is invalid.\
*A snapshot of my code for this is below:*
![CombatCode](https://i.ibb.co/PxSLwLJ/Screen-Shot-2021-04-15-at-9-51-51-PM.png)
After each round, damage is done to the loser's health points based on the winner's attack damage.\
Player's max health points starts at 100 and attack damage starts at 15, but can be increased with potions and by leveling up.\
Enemy's attack damage and health points are generated as a random number.\
Losing all health points in battle will only result in earning no experience, but definite loss in pride.\
Health Points are reset to the Max Health Stat at the beginning of a new encounter.\
Winning a combat sequence results in gaining experience points based on the enemy's attack damage.\
The experince point requirement to level-up starts at 10 points and increases by 5 points each level.

```python
# FUNCTIONS (PLEASE RUN THESE FIRST TO DEFINE)
import random as rd
import time as tm
def gamestart():
    print('\033[1m\033[36m\033[4m------------------------------------WELCOME-TO-VENSHURA------------------------------------\033[0m')
    print("You open your eyes to intensely bright light...\n     \033[95mI can't remember a thing\033[0m")
    tm.sleep(3)
    print("Your eyes quickly adjust to see the entire world around you\n     \033[95mBlissfully unaware of my complications\033[0m")
    tm.sleep(3)
    print("An unfamiliar voice calls out from nearby:")     
    tm.sleep(3)
    print("*Hey you, you're finally awake...*")
    tm.sleep(3)
    print("You turn to find a small, dull-blue, humanoid creature with large, glowing antenna staring up into your eyes..")
    tm.sleep(3)
    print('The creature looked upset.\n     \033[95mAfter a brief pause, they asked me a daunting question:\033[0m')
    tm.sleep(3)
    playername = input("*Can you tell me your name?*: ")
    tm.sleep(2)
    print ("*",playername,", huh?...*")
    tm.sleep(3)
    print("*I suppose that's close enough*")
    tm.sleep(3)
    print("*I guess you really don't remember anything... or what they did to you*")
    tm.sleep(3)
    print("*Maybe that's for the best*")
    tm.sleep(3)
    print("The creature was looking increasingly uncomfortable.\n     \033[95mI wish I could remember something...\033[0m")
    tm.sleep(3)
    print("*You need to get out of here before they find you*")
    tm.sleep(3)
    print("*I'm sorry, but there isn't much time to explain...*")
    tm.sleep(3)
    print("*You are a great warrior.*")
    tm.sleep(3)
    print("*An ancient, evil \033[1m\033[36mBASILISK\033[0m stole your memories*")
    tm.sleep(3)
    print("*You need to follow the path to the \033[1m\033[95mDUNGEON\033[0m.")
    tm.sleep(3)
    print("*The \033[1m\033[36mBASILISK\033[0m must be destroyed*")
    tm.sleep(3)
    print("*Use your \033[1m\033[91mSWORD\033[0m, your \033[1m\033[92mBOW\033[0m, and the \033[1m\033[94mMAGIC\033[0m coursing through your veins to fight your way to the \033[1m\033[95mDUNGEON\033[0m*")
    tm.sleep(3)
    print("*Go my child, there isn't time to waste*")
    return playername

    
def combat(hpstat,attackstat,exp,enemy):
    hp = hpstat
    enemyhp = rd.randint(20,40)
    enemydmg = rd.randint(5,15)
    movecount = 1
    while hp > 0 and enemyhp > 0:
        print ('\033[1m\033[36m\033[4m------------------------------------ ROUND',movecount,'------------------------------------\033[0m\n        | You have\033[1m\033[91m',hp,'HEALTH POINTS\033[0m, and The\033[1m\033[95m',enemy,'\033[0mhas\033[1m\033[91m',enemyhp,'HEALTH POINTS\033[0m | \n\033[1m                     !The\033[95m',enemy,'\033[0m\033[1mis preparing for an attack!\033[0m')
        enemyattack = rd.randint(0,2)
        move = input('Will you attack using \033[1m\033[91m\033[1mMELEE\033[0m, \033[92m\033[1mRANGED\033[0m, or \033[94m\033[1mMAGIC\033[0m?: ')
        if (move == 'MELEE' or move == 'melee' or move == 'Melee') and enemyattack == 0:
            print('\033[95mYour attack is met by an equally strong swing by your opponent...\033[0m')
            tm.sleep(2)
            print("\033[94m\033[1m[YOUR ATTACK: MELEE == ENEMY ATTACK: MELEE]\033[0m\n The clash of your swords briefly stuns the enemy.")
            tie = tiebreaker(attackstat,enemydmg)
            if tie == 2:
                hp -= enemydmg
            elif tie == 1:
                enemyhp -= attackstat
        elif (move == "RANGED" or move == 'ranged' or move == 'Ranged') and enemyattack == 0:
            print('\033[95mYou sent an arrow down the path and it reaches your enemy before they even ready their weapon...\033[0m')
            tm.sleep(2)
            print("\033[92m\033[1m[YOUR ATTACK: RANGED => ENEMY ATTACK: MELEE]\033[0m\nThe enemy lost\033[1m\033[91m",attackstat,'HEALTH POINTS\033[0m')
            enemyhp -= attackstat
        elif (move == "MAGIC" or move == 'magic' or move == 'Magic') and enemyattack == 0:
            print('\033[95mYour attempt to cast a fireball at your opponent was dodged and met with a strong swift punch...\033[0m')
            tm.sleep(2)
            print("\033[93m\033[1m[YOUR ATTACK: MAGIC <= ENEMY ATTACK: MELEE]\033[0m\nYou lost\033[1m\033[91m",enemydmg,'HEALTH POINTS\033[0m')
            hp -= enemydmg
        elif (move == 'MELEE' or move == 'melee' or move == 'Melee') and enemyattack == 1:
            print('\033[95mYour enemy hurled a boulder in your direction while you fumbled with your sword...\033[0m')
            tm.sleep(2)
            print("\033[93m\033[1m[YOUR ATTACK: MELEE <= ENEMY ATTACK: RANGED]\033[0m\nYou lost\033[1m\033[91m",enemydmg,'HEALTH POINTS\033[0m')
            hp -= enemydmg
        elif (move == "RANGED" or move == 'ranged' or move == 'Ranged') and enemyattack == 1:
            print('\033[95mYou ready an arrow in on your bow and notice your foe doing the same...\033[0m')
            tm.sleep(2)
            print("\033[94m\033[1m[YOUR ATTACK: RANGED == ENEMY ATTACK: RANGED]\033[0m\n A mid air collision of your ranged attacks distracts your enemy.")
            tie = tiebreaker(attackstat,enemydmg)
            if tie == 2:
                hp -= enemydmg
            elif tie == 1:
                enemyhp -= attackstat
        elif (move == "MAGIC" or move == 'magic' or move == 'Magic') and enemyattack == 1:
            print('\033[95mBefore your opponent can ready their bow, you cast a lightning bolt down at their position...\033[0m')
            tm.sleep(2)
            print("\033[92m\033[1m[YOUR ATTACK: MAGIC => ENEMY ATTACK: RANGED]\033[0m\nThe enemy lost\033[1m\033[91m",attackstat,'HEALTH POINTS\033[0m')
            enemyhp -= attackstat
        elif (move == 'MELEE' or move == 'melee' or move == 'Melee') and enemyattack == 2:
            print("\033[95mYour opponent's attempt at sorcery was quelched with a quick and brutal swing from your sword...\033[0m")
            tm.sleep(2)
            print("\033[92m\033[1m[YOUR ATTACK: MELEE => ENEMY ATTACK: MAGIC]\033[0m\nThe enemy lost\033[1m\033[91m",attackstat,'HEALTH POINTS\033[0m')
            enemyhp -= attackstat
        elif (move == "RANGED" or move == 'ranged' or move == 'Ranged') and enemyattack == 2:
            print('\033[95mYour opponent managed to send a primitive fireball your way before you could aim your bow...\033[0m')
            tm.sleep(2)
            print("\033[93m\033[1m[YOUR ATTACK: RANGED <= ENEMY ATTACK: MAGIC]\033[0m\nYou lost\033[1m\033[91m",enemydmg,'HEALTH POINTS\033[0m')
            hp -= enemydmg
        elif (move == "MAGIC" or move == 'magic' or move == 'Magic') and enemyattack == 2:
            print('\033[95mJust as you managed to create an ice shard from thin air, your enemy mustered an ethereal ball of lightning...\033[0m')
            tm.sleep(2)
            print("\033[94m\033[1m[YOUR ATTACK: MAGIC == ENEMY ATTACK: MAGIC]\033[0m\n A large explosion from the combination of your magic attacks knocks your foe off balance.")
            tie = tiebreaker(attackstat,enemydmg)
            if tie == 2:
                hp -= enemydmg
            elif tie == 1:
                enemyhp -= attackstat
        else:
            print("\033[1mINVALID SELECTION...\033[0m")
            continue
        movecount += 1
        if hp <= 0:
            print('You have fallen in battle after',movecount,'rounds...')
            print('\033[1m\033[36m\033[4m----------------------------------------------------------------------------------\033[0m\n')
            return 0
        elif enemyhp <= 0:
            print('After',movecount,'rounds, you have slain the\033[1m\033[95m',enemy,'\033[0mand gained\033[1m\033[95m',enemydmg,"EXPERIENCE POINTS\033[0m")
            exp += enemydmg
            print('\033[1m\033[36m\033[4m----------------------------------------------------------------------------------\033[0m\n')
            return exp
        

def tiebreaker(attackstat,enemydmg):
    counter = 0
    enemystun = rd.randint(0,1)
    while (counter != 'COUNTER' or counter != 'counter' or counter != 'Counter' or counter != 'GUARD' or counter != 'guard' or counter != 'Guard'):
        counter = input('Will you \033[1m\033[95mCOUNTER\033[0m or \033[1m\033[93mGUARD\033[0m?: ')
        if (counter == 'GUARD' or counter == 'guard' or counter == 'Guard') and enemystun == 1:
            print('\033[95mYou followed through with a strong guard, but your enemy remained stunned.\033[0m')
            tm.sleep(2)
            print('No one was harmed.')
            return 0
        elif (counter == 'COUNTER' or counter == 'counter' or counter == 'Counter') and enemystun == 1:
            print('\033[95mYou caught your opponent off guard and made him regret it.\033[0m')
            tm.sleep(2)
            print('The enemy lost\033[1m\033[91m',attackstat,'HEALTH POINTS\033[0m')
            return 1
        elif (counter == 'GUARD' or counter == 'guard' or counter == 'Guard') and enemystun == 0:
            print('\033[95mYour enemy snaps out of it and resumes an attack-ready stance.\033[0m')
            tm.sleep(2)
            print('No one was harmed.')
            return 0
        elif (counter == 'COUNTER' or counter == 'counter' or counter == 'Counter') and enemystun == 0:
            print("\033[95mYour foe's stun faded before you could land the counter-attack and they deflected.\033[0m")
            tm.sleep(2)
            print("You lost\033[1m\033[91m",enemydmg,'HEALTH POINTS\033[0m')
            return 2
        else:
            print("\033[1mINVALID SELECTION...\033[0m")


def gift():
    race = ['ELVEN','DWARVISH','ORCISH','GNOMISH','HUMAN','FAE']
    job = ['MERCHANT','TRADER','ALCHEMIST','ADVENTURER','HUNTER','FARMER']
    npcrace = race[rd.randint(0,5)]
    npcjob = job[rd.randint(0,5)]
    print('\033[1m\033[92m\033[4m-------------------------------FRIENDLY-ENCOUNTER-------------------------------\033[0m')
    print('While walking the path, a friendly\033[94m\033[1m', npcrace ,npcjob, '\033[0mseems to recognize me....\n     \033[95mThey offered me a free potion!\033[0m')
    pickup = input('Will you take the \033[1m\033[91mHEALTH POTION\033[0m, \033[1m\033[94mATTACK POTION\033[0m, or \033[1m\033[95mEXPERIENCE POTION\033[0m?: ')
    if pickup == 'HEALTH' or pickup == 'health' or pickup == 'Health' or pickup == 'HEALTH POTION' or pickup == 'health potion' or pickup == 'Health Potion' or pickup == 'Health potion':
        pickup = 0
        print('The\033[94m\033[1m', npcrace ,npcjob, '\033[m graciously handed you a bottle and continued the opposite way the trail...')
    elif pickup == 'ATTACK' or pickup == 'attack' or pickup == 'Attack' or pickup == 'ATTACK POTION' or pickup == 'attack potion' or pickup == 'Attack Potion' or pickup == 'Attack potion':
        pickup = 1
        print('The\033[94m\033[1m', npcrace ,npcjob, '\033[m graciously handed you a bottle and continued the opposite way the trail...')
    elif pickup == 'EXPERIENCE' or pickup == 'experience' or pickup == 'Experience' or pickup == 'EXPERIENCE POTION' or pickup == 'experience potion' or pickup == 'Experience Potion' or pickup == 'Experience potion':
        pickup = 2
        print('The\033[94m\033[1m', npcrace ,npcjob, '\033[m graciously handed you a bottle and continued the opposite way the trail...')
    else:
        pickup = 3
        print('The\033[94m\033[1m', npcrace ,npcjob, '\033[mcould not understand what you wanted and stormed off, completely offended.')
    return pickup


def dungeon(playername,hpstat):
    wordlist = ['FLAME','SWORD','WATER','STAFF','FIERY','ARMOR','LIGHT','POWER','STORM','EARTH','CHAOS','FROST']
    word = wordlist[rd.randint(0,11)]
    guessed = []
    hp = hpstat
    roundcnt = 0
    tm.sleep(3)
    print("Suddenly the path you walk on spills out of the thick woods and reveals a massive structure...\n     \033[95mThis has to be it.\033[0m")
    tm.sleep(3)
    print("With your approach, large stone doors crash open from otherworldly power...\n     \033[95mI must be expected.\033[0m")
    tm.sleep(3)
    print("Ancient carvings on the wall depict a massive serpent destroying peaceful villages of old.")
    tm.sleep(3)
    print("The heiroglyphs describe a single-worded spell which can be used to completely destroy it...")
    tm.sleep(3)
    print("...The word seems to have been worn away by incisions from large fangs.\n     \033[95mOf course...\033[0m")
    tm.sleep(3)
    print("As you trek deeper into the cavernous hallways, torches illuminate a dead end ahead...\n     \033[95mIt's a trap.\033[0m")
    tm.sleep(3)
    print('The \033[1m\033[36mBASILISK\033[0m wraps his tail around your neck and tries to dangle you from the ceiling.')
    tm.sleep(3)
    print("*WHY, IF IT ISSSSN'T",playername,'?*')
    tm.sleep(3)
    print('The ancient evil creature spoke with a deep hiss and complete hatred behind his words:')
    retort = input('*DID YOU REALLY EXXXXPECT TO COME HERE AND DESSSSTROY ME?*')
    print("*SSSSHUTUP MORTAL*")
    tm.sleep(2)
    print("The serpent king's voice echoed throughout the dungeon\n     \033[95mThis is my only chance, I need to fight! Maybe I can figure out that word...\033[0m")
    tm.sleep(3)
    print('\033[1m\033[91m------------------------------THE-FINAL-BATTLE------------------------------\033[0m')
    tm.sleep(2)
    print("*IT'SSSS A SSSSHAME YOU TRIED SSSSO HARD JUSSSST TO DIE*\n     \033[95m*I was going to say the same to you*\033[0m")
    while hp > 0:
        roundcnt += 1
        display = ''
        hits = 0
        print('\033[1m\033[91m\033[4m----------------------------------ROUND',roundcnt,'----------------------------------\033[0m\n                            | You have\033[1m\033[91m',hp,'HEALTH POINTS\033[0m |        ')
        guess = input('Guess a letter \033[1m(IN UPPERCASE)\033[0m to try and defeat the \033[1m\033[36mBASILISK\033[0m!: ')
        print('You swing for the \033[1m\033[36mBASILISK\033[0m...')
        tm.sleep(1)
        if guess in guessed:
            print("But your attack is inefective...")
            tm.sleep(1)
            print("You've already tried",guess,'. Try another letter.')
            continue
        elif guess in word:
            print('Your land an attack and damage the \033[1m\033[36mBASILISK\033[0m!')
            guessed.append(guess)
            tm.sleep(2)
        else:
            print('...But miss')
            hp -= 10
            print('The \033[1m\033[36mBASILISK\033[0m tightens its grip on your neck...\nYou lost \033[91m\033[1m10 HEALTH POINTS\033[0m!')
            guessed.append(guess)
            tm.sleep(2)
        for i in word:
            if i in guessed:
                display += i
                hits += 1
            else:
                display += '-'
        if display == word:
            print("You have solved for the word:\033[1m",word,"\033[0mand defeated the \033[1m\033[36mBASILISK\033[0m with an incredible blast of pure arcane energy!")
            tm.sleep(3)
            print("\033[92m\033[1m\033[4m                    YOU REMEMBER EVERYTHING                    \033[0m")
            tm.sleep(3)
            print('*You have completed Venshura...*\n     \033[1m\033[95mThank you for playing!\033[0m')
            break
        tm.sleep(2)
        print('\033[1m\033[36mSOLVE:',display,'\033[0m')
        print('\033[1m\033[92mAready Guessed:',guessed,'\033[0m')
        tm.sleep(1)
    if hp <= 0:
        print("You have fallen in battle...\n      \033[95mI need to find my strength and attack again!\033[0m")
        tm.sleep(2)
        dungeon(playername,hpstat)
    
```
